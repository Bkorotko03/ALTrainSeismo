```python
import numpy as np
import pandas as pd

readVoltage = pd.read_csv('./output/overnight_run_apr1_voltage.csv')
readPassby = pd.read_csv('./testing.csv')

voltage_time_col = 'seconds'
passby_time_col = 'seconds'
voltage_col = 'voltage'

half_window = 500
max_time_diff = None

readVoltage[voltage_time_col] = pd.to_numeric(readVoltage[voltage_time_col], errors='coerce')
readPassby[passby_time_col] = pd.to_numeric(readPassby[passby_time_col], errors='coerce')
readVoltage[voltage_col] = pd.to_numeric(readVoltage[voltage_col], errors='coerce')

# shift voltage time so it also starts at 0
readVoltage[voltage_time_col] = readVoltage[voltage_time_col] - 1168

readVoltage = readVoltage.dropna(subset=[voltage_time_col, voltage_col]).copy()
readPassby = readPassby.dropna(subset=[passby_time_col]).copy()

readVoltage[voltage_col] = readVoltage[voltage_col].fillna(0)

readVoltage = readVoltage.sort_values(voltage_time_col).reset_index(drop=True)
readPassby = readPassby.sort_values(passby_time_col).reset_index(drop=True)

voltage_times = readVoltage[voltage_time_col].to_numpy()
passby_times = readPassby[passby_time_col].to_numpy()

in_window = np.zeros(len(readVoltage), dtype=bool)
match_report = []

for passby_index, t in enumerate(passby_times):
    match_index = np.argmin(np.abs(voltage_times - t))
    match_time = voltage_times[match_index]
    time_diff = abs(match_time - t)

    accepted = True
    if max_time_diff is not None and time_diff > max_time_diff:
        accepted = False

    match_report.append({
        'passby_index': passby_index,
        'passby_time': t,
        'matched_v_index': match_index,
        'matched_v_time': match_time,
        'time_diff': time_diff,
        'accepted': accepted
    })

    if not accepted:
        continue

    start_index = max(0, match_index - half_window)
    end_index = min(len(readVoltage), match_index + half_window + 1)
    in_window[start_index:end_index] = True

matchReport = pd.DataFrame(match_report)

eventwindowspassby = readVoltage.copy()
eventwindowspassby.loc[~in_window, voltage_col] = 0

backgroundwindowspassby = readVoltage.copy()
backgroundwindowspassby.loc[in_window, voltage_col] = 0

events_output_passby_file = 'eventspassby.csv'
background_output_passby_file = 'backgroundpassby.csv'

eventwindowspassby.to_csv(events_output_passby_file, index=False)
backgroundwindowspassby.to_csv(background_output_passby_file, index=False)
matchReport.to_csv('match_report.csv', index=False)

print("Done.")
print()
print(f'Rows in the csv from arduino: {len(readVoltage)}')
print(f'Rows in csv made from our clicker data: {len(readPassby)}')
print(f'Accepted passby matches: {matchReport["accepted"].sum()} / {len(matchReport)}')
print(f'Rows inside event windows: {in_window.sum()}')
print(f'Rows outside event windows: {(~in_window).sum()}')
print()
print(f'Saved: {events_output_passby_file}')
print(f'Saved: {background_output_passby_file}')
print('Saved: match_report.csv')

nonzero_eventspassby = (eventwindowspassby[voltage_col] != 0).sum()
nonzero_backgroundpassby = (backgroundwindowspassby[voltage_col] != 0).sum()

print()
print(f'Nonzero voltages in eventspassby.csv: {nonzero_eventspassby}')
print(f'Nonzero voltages in backgroundpassby.csv: {nonzero_backgroundpassby}')
print(f'Total rows check: {len(eventwindowspassby)} and {len(backgroundwindowspassby)}')
```
